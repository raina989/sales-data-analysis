```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("../data/train.csv")

# View first rows
df.head()
```

```python
# Region with highest sales
```

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values()

plt.figure(figsize=(8,5))
region_sales.plot(kind="barh", title="Sales by Region")

plt.tight_layout()

# SAVE THE IMAGE
plt.savefig("../visuals/sales_by_region.png")

plt.show()
```

```python
# most profitable category
```

```python
region_sales = df.groupby("Category")["Sales"].sum().sort_values()

plt.figure(figsize=(8,5))
region_sales.plot(kind="barh", title="Sales by Category")

plt.tight_layout()

# SAVE THE IMAGE
plt.savefig("../visuals/sales_by_category.png")

plt.show()
```

```python
# monthly sales trend
```

```python
region_sales = df.groupby("Order Date")["Sales"].sum().sort_values()

plt.figure(figsize=(8,5))
region_sales.plot(kind="line", title="Monthly Sales")

plt.tight_layout()

# SAVE THE IMAGE
plt.savefig("../visuals/monthly_sales.png")

plt.show()
```

```python
Top-selling products
```

```python
product_sales = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,15))
product_sales.head(20).plot(kind="barh", title="Top 20 Products by Sales")

plt.tight_layout()

# SAVE THE IMAGE
plt.savefig("../visuals/Top-20-Products-by-Sales.png")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=30)

plt.title("Distribution of Sales")
plt.xlabel("Sales Amount")
plt.ylabel("Number of Orders")

plt.tight_layout()
plt.savefig("../visuals/sales_distribution.png")
plt.show()
```
